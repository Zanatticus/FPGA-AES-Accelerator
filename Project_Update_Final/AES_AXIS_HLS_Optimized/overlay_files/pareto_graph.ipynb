```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Define the x and y coordinates for the points
latency = [11828, 10386, 10334, 9753, 8665, 8595, 8057, 8025, 7973, 7379]
BRAM = [9, 9, 9, 9, 9, 8, 8, 8, 8, 9]
FF = [1785, 4405, 4555, 4548, 4811, 4929, 4990, 5003, 5018, 5169]
LUT = [7806, 10751, 11235, 11516, 12256, 12711, 12703, 12770, 13011, 13126]
area = [100*(BRAM[i]) + max(FF[i],LUT[i]) for i in range(10)]

# Define the colors for the points
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'black', 'gray']
# Plot the points
plt.scatter(latency, area, c=colors)
plt.title('Pareto Graph: HLS Optimizations')
plt.xlabel('Latency (ns)')
plt.ylabel('Area')

label_list = ["Baseline", "expandKey", "subBytes", "createRoundKey", "aes_main", "invSubBytes", "invShiftRows", "shiftRow/shiftRows", "invMixColumns", "aes_invMain"]
legend_elements = [Line2D([0], [0], marker='o', color='w', label=label_list[i], markerfacecolor=color, markersize=10) for i, color in enumerate(colors)]
plt.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.45, 1))

# Show the plot
plt.show()
```
